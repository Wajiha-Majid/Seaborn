```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({
'Value': np.random.normal(0, 1, 1000)
})
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.Figure(figsize=(9,4))
x = np.linspace(0, 10, 100)
w=np.sin(x)
v=np.cos(x)
plt.subplot(221)
plt.plot(x,v,w)
plt.subplot(222)
colors=np.random.rand(100)
z = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(z,y, cmap='coolwarm')
plt.subplot(223)
plt.bar(categories,values, facecolor='red')
plt.subplot(224)
sns.histplot(data, bins=20, kde=True )
plt.tight_layout(pad=0.2)
plt.show()
```
