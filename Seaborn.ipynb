```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
iris= sns.load_dataset('iris')
mpg= sns.load_dataset('mpg')
tips= sns.load_dataset('tips')
```

```python
dat= np.random.rand(5, 5)
sns.heatmap(dat,annot=True,cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(x='sepal_length', y='sepal_width', data=iris)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.histplot(iris['sepal_length'],kde=True)
plt.show()
sns.histplot(tips['size'],kde=True)
plt.show()
```

```python
num=mpg.select_dtypes(include='number')
corr=num.corr()
sns.heatmap(corr, annot=True, cmap='viridis')
plt.show()
```

```python
num=tips.select_dtypes(include='number')
corr=num.corr()
sns.heatmap(corr, annot=True, cmap='viridis')
plt.show()
```

```python
sns.set_palette('bright')
sns.set_style('ticks')
sns.boxplot(x='size', y='total_bill', data=tips)
plt.show()
```

```python
sns.set_context('poster')
sns.scatterplot(x='total_bill', y='size', data=tips)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='pastel')
plt.show()
```
